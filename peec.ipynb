```python
def fibonacci(n):
    # return a list of fibonacci numbers
    fibonacci_lst = []
    def find_fibonacci(n):
        if n == 0:
            return 0
        elif n == 1:
            return 1
        else:
            return find_fibonacci(n-2) + find_fibonacci(n-1)
    for _ in range(n):
        fibonacci_lst.append(find_fibonacci(n))
    return fibonacci_lst
```

```python
import pandas as pd

df = pd.DataFrame({
    "product_id": [101, 101, 101, 102, 102, 102, 102],
    "date": [
        "2023-01-01", "2023-01-02", "2023-01-03",
        "2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04"
    ],
    "units_sold": [10, 15, 20, 5, 7, 9, 11]
})

df['date'] = pd.to_datetime(df['date'])
```

```python
df
```

```python


# Sort values (important for rolling!)
df = df.sort_values(['product_id', 'date'])

# Group by product, set date as index so rolling works
df['rolling_avg_7d'] = (
    df
    .set_index('date')                                  # step 1
    .groupby('product_id')['units_sold']                # step 2
    .rolling('7D', min_periods=1)                       # 7-day window
    .mean()
    .reset_index(level=0, drop=True)                    # cleanup
)

# Optional: reset index if needed
df = df.reset_index(drop=True)
```
